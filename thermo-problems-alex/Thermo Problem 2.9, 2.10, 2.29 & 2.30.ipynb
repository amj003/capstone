## Problem 2.9 

Use a computer to reproduce the table and graph in Figure 2.4: two Einstein solids, each containing three harmonic oscillators, with a total of six units of energy. Then modify the table and graph to show the case where one Einstein solid contains six harmonic oscillators and the other contains four harmonic oscillators (with a total number of energy units still equal to six). Assuming that all microstates are equally likely, what is the most probable macrostate, and what is it's probability?

```python
import pandas as pd
import numpy as np
import math

%pylab inline

```

```python
def mults(N_A, N_B, totalenergy):
    q_A = list(xrange(totalenergy+1))
    q_B = list(range(totalenergy, -1, -1))
    omega_A = list(float(math.factorial(N_A + q_A[n] - 1)/(math.factorial(q_A[n])*math.factorial(N_A-1))) for n in q_A)
    omega_B = list(float(math.factorial(N_B + q_B[n] - 1)/(math.factorial(q_B[n])*math.factorial(N_B-1))) for n in q_A)
    omegatotal = list(float(omega_A[n]*omega_B[n]) for n in q_A)
    probability = list(float(omegatotal[n])/float(sum(omegatotal)) for n in q_A)
    
    data = pd.DataFrame({'\r$\Omega_B$':omega_B,
                         '\r$q_B$':q_B,
                         '\r$q_A$':q_A,
                         '\r$\Omega_A$':omega_A,
                         '\r$\Omega_{Total}$':omegatotal,
                         '\rProbability':probability})

    return data 

data = mults(3,3,6)
data[['\r$q_A$', '\r$\Omega_A$','\r$q_B$','\r$\Omega_B$','\r$\Omega_{Total}$', '\rProbability']]
```

The sum of $\Omega_{total} = \Omega_A \Omega_B$ is 462, from $$\Omega(N,q) = {{(q + N - 1)!} \over {q! (N - 1)!}} = {{(6 + 6 - 1)!} \over {6! (6 - 1)!}} = {{11!} \over {6! 5!}} = {{(11)(10)(9)(8)(7)} \over {(5)(4)(3)(2)}} = (11)(7)(2)(3) = 462$$

```python

print 'The total sum of the microstates is', data['\r$\Omega_{Total}$'].sum(), '.'

def plots(dataframe, ptype='none' ):

    if ptype == 'none':
        plt.figure(1)
        dataframe.groupby('\r$q_A$')['\r$\Omega_{Total}$'].sum().plot(figsize = (9,6))
        plt.xlabel('$q_A$', fontsize = 20)
        plt.ylabel('$\Omega_{Total}$',fontsize = 20)


        plt.figure(2)
        dataframe.groupby('\r$q_A$')['\rProbability'].sum().plot(figsize = (9,6), color = 'g')
        plt.xlabel('$q_A$', fontsize = 20)
        plt.ylabel('Probability',fontsize = 20)
    else: 
        plt.figure(1)
        dataframe.groupby('\r$q_A$')['\r$\Omega_{Total}$'].sum().plot(kind=ptype, figsize = (9,6))
        plt.xlabel('$q_A$', fontsize = 20)
        plt.ylabel('$\Omega_{Total}$',fontsize = 20)


        plt.figure(2)
        dataframe.groupby('\r$q_A$')['\rProbability'].sum().plot(kind=ptype, figsize = (9,6), color = 'g')
        plt.xlabel('$q_A$', fontsize = 20)
        plt.ylabel('Probability',fontsize = 20)
    
plots(data, 'bar')
```

```python
data2 = mults(6,4,6)

print 'The total sum of the microstates is', data2['\r$\Omega_{Total}$'].sum(), '.'

data2[['\r$q_A$', '\r$\Omega_A$','\r$q_B$','\r$\Omega_B$','\r$\Omega_{Total}$','\rProbability']]
```

The sum of $\Omega_{total} = \Omega_A \Omega_B$ is 5005, from $$\Omega(N,q) = {{(q + N - 1)!} \over {q! (N - 1)!}} = {{(6 + 10 - 1)!} \over {6! (10 - 1)!}} = {{15!} \over {6! 9!}} = {{(15)(14)(13)(12)(11)(10)} \over {(6)(5)(4)(3)(2)}} = {{(14)(13)(11)(10)} \over {(4)}} = 5005$$

```python
plots(data2, 'bar')
```

The most probable microstate is where $q_A = 4$, which has a probability of $ {1260 \over 5005}*100$ which comes out to roughly 25%. 

# Problem 2.10

Use a computer to produce the table and graph for the case where one Einstein solid contains 200 oscillators, the other contains 100 oscillators, and there are 100 units of energy total. What is the most probable macrostate, and what is it's probability? What is the least probable macrostate, and what is it's probability?

```python
data3 = mults(200,100,100)

data3[['\r$q_A$', '\r$\Omega_A$','\r$q_B$','\r$\Omega_B$','\r$\Omega_{Total}$','\rProbability']]

```

```python
plots(data3)

for n in data3['\r$q_A$']:
    if data3['\rProbability'][n] == data3['\rProbability'].max():
        maxprob = data3['\r$q_A$'][n]
    elif data3['\rProbability'][n] == data3['\rProbability'].min():
        minprob = data3['\r$q_A$'][n]
        
print 'The most probable microstate is', data3['\r$q_A$'][maxprob], 'units of energy in solid A. This state has a probability of',100*data3['\rProbability'].max(), '%.'

print 'The least probable microstate is', data3['\r$q_A$'][minprob], 'units of energy in solid A. This state has a probability of',100*data3['\rProbability'].min(), '%.'
```
