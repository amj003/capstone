# Lagrangian

```python
import numpy as np
from sympy import *
from sympy.physics.vector import dynamicsymbols
init_printing()
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from scipy.integrate import odeint

#Let l = 1, m = 1, g = 1 
```

```python
#Use dynamicsymbols to make theta and phi depend on time.
theta = dynamicsymbols('theta')
phi = dynamicsymbols('phi')
d_theta = dynamicsymbols('d_theta')
d_phi = dynamicsymbols('d_phi')
t = Symbol('t')
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')
d_x = Symbol('d_x')
d_y = Symbol('d_y')
d_z = Symbol('d_z')
L = Symbol('L')
w = Symbol('w')
F = Symbol('F')
b = Symbol('b')
```

```python
x = sin(theta)*cos(phi)
y = sin(theta)*sin(phi)
z = -cos(theta) 
#+ F*cos(w*t)

#Define d_theta and d_phi

d_theta = diff(theta, t)
d_phi = diff(phi, t)

#Define the derivatives of the coordinate system 

d_x = diff(x,t)
d_y = diff(y,t)
d_z = diff(z,t)

#Find the velocity of the pendulum

v = simplify(expand((d_x)**2 + (d_y)**2 + (d_z)**2))

#Find kinetic energy and potential energy. Kinetic is T = 1/2*m*v**2 and 
#potential is U = mgh where h is just z_1.
#The driving force is already in z_1.
#Define the lagrangian.

T = 0.5*v
U = z
L = T - U

L
```

```python

L_d_theta = diff(L, theta)
L_dd_theta = diff(L, d_theta)
eq_theta = diff(L_dd_theta, t) - L_d_theta

L_d_phi = diff(L, phi)
L_dd_phi = diff(L, d_phi)
eq_phi = diff(L_dd_phi, t) - L_d_phi

solve([eq_theta, eq_phi], [diff(d_theta,t), diff(d_phi,t)])
```

```python
#theta'' = sin(theta)*cos(theta)*phi'**2 - sin(theta)
#phi'' = (-2*phi'*theta')/tan(theta)

#u1(t) = theta(t)
#u2(t) = theta'(t)
#u1'(t) = u2(t)
#u2'(t) = sin(u1)*cos(u1)*w2**2 - sin(u1)

#w1(t) = phi(t)
#w2(t) = phi'(t)
#w1'(t) = w2(t)
#w2'(t) = (-2*w2*u2)/tan(u1)

def f(vec, t):
    u1, w1, u2, w2 = vec
    return [u2, 
            w2, 
            sin(u1)*cos(u1)*w2**2 - sin(u1),
            (-2*w2*u2)/tan(u1)]

vec = [pi/2, 0.1, 0.0, 0.1]
time = np.linspace(0, 100, 1000)
time_list = xrange(0, 1000)
results = odeint(f, vec, time)


theta_f = list(results[n,0] for n in time_list)
phi_f = list(results[n,1] for n in time_list)
theta_dot_f = list(results[n,2] for n in time_list)
phi_dot_f = list(results[n,3] for n in time_list)

sin_theta = list(sin(x) for x in theta_f)
cos_phi = list(cos(x) for x in phi_f)
cos_theta = list(cos(x) for x in theta_f)
sin_phi = list(sin(x) for x in phi_f)


x_f = list(sin(theta_f[i])*cos(phi_f[i]) for i in time_list)
y_f = list(sin(theta_f[i])*sin(phi_f[i]) for i in time_list)
z_f = list(-cos(theta_f[i]) for i in time_list)

```

```python
fig = pylab.figure()
ax = Axes3D(fig)
ax.plot(x_f, y_f,z_f)
pylab.savefig('undampedspherical3d.png')

```

```python
#theta'' = F*cos(t*w) - sin(theta) + 0.5*sin(2*theta)*phi'**2
#phi'' = (-2*phi'*theta')/tan(theta)

#u1(t) = theta(t)
#u2(t) = theta'(t)
#u1'(t) = u2(t)
#u2'(t) = F*cos(t*w) - sin(u1) + 0.5*sin(2*u1)*w2**2

#w1(t) = phi(t)
#w2(t) = phi'(t)
#w1'(t) = w2(t)
#w2'(t) = (-2*w2*u2)/tan(u1)

def f2(vec, t):
    w = 1.0
    F = 1.0
    u1, w1, u2, w2 = vec
    return [u2, 
            w2, 
            F*cos(t*w) - sin(u1) + 0.5*sin(2*u1)*w2**2, 
            (-2*w2*u2)/tan(u1)]

vec2 = [0.5, 0.5, 0.5, 0.5]
time2 = np.linspace(0, 100, 1000)
time_list2 = xrange(0, 1000)
results2 = odeint(f2, vec2, time2)


theta2_f = list(results2[n,0] for n in time_list2)
phi2_f = list(results2[n,1] for n in time_list2)
theta_dot2_f = list(results2[n,2] for n in time_list2)
phi_dot2_f = list(results2[n,3] for n in time_list2)

sin_theta2 = list(sin(x) for x in theta2_f)
cos_phi2 = list(cos(x) for x in phi2_f)
cos_theta2 = list(cos(x) for x in theta2_f)
sin_phi2 = list(sin(x) for x in phi2_f)

x2_f = list(sin(theta2_f[i])*cos(phi2_f[i]) for i in time_list2)
y2_f = list(sin(theta2_f[i])*sin(phi2_f[i]) for i in time_list2)
z2_f = list(-cos(theta2_f[i]) for i in time_list2)
```
