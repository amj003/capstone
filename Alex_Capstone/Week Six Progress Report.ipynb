```python
import numpy as np
from sympy import *
from sympy.physics.vector import dynamicsymbols
init_printing()
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from scipy.integrate import odeint
np.set_printoptions(threshold=np.inf)


#Let l = 1, m = 1, g = 1 
```

```python
theta_1 = dynamicsymbols('theta_1')
d_theta_1 = dynamicsymbols('d_theta_1')
theta_2 = dynamicsymbols('theta_2')
d_theta_2 = dynamicsymbols('d_theta_2')
t = Symbol('t')
x_1 = Symbol('x_1')
y_1 = Symbol('y_1')
d_x_1 = Symbol('d_x_1')
d_y_1 = Symbol('d_y_1')
x_2 = Symbol('x_2')
y_2 = Symbol('y_2')
d_x_2 = Symbol('d_x_2')
d_y_2 = Symbol('d_y_2')
L_1 = Symbol('L_1')
L_2 = Symbol('L_2')
w = Symbol('w')
F = Symbol('F')
b = Symbol('b')
l = Symbol('l')
m = Symbol('m')
g = Symbol('g')
```

```python
d_theta_1 = diff(theta_1, t)
d_theta_2 = diff(theta_2, t)

x_1 = l*sin(theta_1)
y_1 = -l*cos(theta_1)
x_2 = l*sin(theta_1) + l*sin(theta_2)
y_2 = -l*cos(theta_1) - l*cos(theta_2)

v_1 = diff(x_1,t)**2 + diff(y_1,t)**2 

L_1 = 0.5*m*v_1  - m*g*y_1

v_2 = diff(x_2,t)**2 + diff(y_2,t)**2

L_2 = 0.5*m*(v_1 + v_2) - m*g*(y_1 + y_2)

expand(simplify(L_2)) 
```

```python
#The Lagrangian for the planar double pendulum is: 

L_2
```

```python

L_d_theta = diff(L, theta)
L_dd_theta = diff(L, d_theta)
eq_theta = diff(L_dd_theta, t) - L_d_theta

L_d_phi = diff(L, phi)
L_dd_phi = diff(L, d_phi)
eq_phi = diff(L_dd_phi, t) - L_d_phi

solve([eq_theta, eq_phi], [diff(d_theta,t), diff(d_phi,t)])

eq_theta, eq_phi

```

```python
L_d_theta_1 = diff(L_2, theta_1)
L_dd_theta_1 = diff(L_2, d_theta_1)
eq_theta_1 = diff(L_dd_theta_1, t) - L_d_theta_1

expand(simplify(eq_theta_1))

L_d_theta_2 = diff(L_2, theta_2)
L_dd_theta_2 = diff(L_2, d_theta_2)
eq_theta_2 = diff(L_dd_theta_2, t) - L_d_theta_2

expand(simplify(eq_theta_2))
```

```python
#We get out two equations of motion we can solve for the theta 1 double dot and theta 2 double dot. 

expand(simplify(eq_theta_1)), expand(simplify(eq_theta_2))
```
