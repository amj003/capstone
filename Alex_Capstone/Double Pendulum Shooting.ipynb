# Section One: Augjac

Step 1: Define the "$2n$-dimensional state vector", as Peeters says, by creating 

$$\mathbf{z} = \left(\begin{array}{c} \theta_1 \\ \theta_2 \\ \dot{\theta_1} \\ \dot{\theta_2} \end{array} \right). $$

To do this we use the same matrix method as when we were focusing on plotting the motion of the double pendulum in physical space. Note for ease of use I have set all paramaters to $1.0$ but the code may be modified to make these passed in values. 

```python
def zdot(z, t):
    x1, x2, x1dot, x2dot = z
    x1ddot = -0.5*x1**3-(2*x1-x2)
    x2ddot = -(2*x2-x1)
    return (x1dot, x2dot, x1ddot, x2ddot)

def Etot(xvec):
    x1, x2, x1dot, x2dot = xvec
    kinetic = (x1dot**2 + x2dot**2)
    potential = x1**2 + x2**2 - x1*x2 + x1**4/2.0
    return kinetic + potential

print "Eigenvector 1: [-1.0, 1.0, 0.0, 0.0], omega = %f, T= %f" % (np.sqrt(3),2*np.pi/np.sqrt(3))
print "Eigenvector 2: [1.0, 1.0, 0.0, 0.0], omega = %f, T= %f" % (np.sqrt(1),2*np.pi/np.sqrt(1))
```

```python
def zdot(z, t):
    theta1, theta2, theta1dot, theta2dot = z
    m1 = 1.0
    m2 = 1.0
    l1 = 1.0
    l2 = 1.0
    g = 1.0
    
    M = np.matrix([[ (m1+m2)*l1**2, l1*l2*m2*np.cos(theta1-theta2)], 
                   [l1*l2*m2*np.cos(theta1-theta2), l2**2*m2]])

    B = np.matrix([[-g*l1*m1*np.sin(theta1) - g*l1*m2*np.sin(theta1) - l1*l2*m2*np.sin(theta1 - theta2)*theta2dot**2], 
                   [-g*l2*m2*np.sin(theta2) + l1*l2*m2*np.sin(theta1-theta2)*theta1dot**2]])

    
    resu = M.I*B
    theta1ddot = resu[0,0]
    theta2ddot = resu[1,0]
    return (theta1dot, theta2dot, theta1ddot, theta2ddot)

#given some z = [theta_1, theta_2, d_theta_1, d_theta_2]

def Etot(z):
    return 2.*(1. - np.cos(z[0])) + (1. - np.cos(z[1])) + z[2]**2 + 0.5*z[3]**2. + np.cos(z[0] - z[1])*z[2]*z[3]

print "Eigenvector 1: [−0.707106781187, 1.0, 0.0, 0.0], omega = 1.847759, T= 3.400435"
print "Eigenvector 2: [0.707106781187, 1.0, 0.0, 0.0], omega = 0.765367, T= 8.209377"
```

Define the firing function (since it's shooting method... ). This is just the time integration of $\mathbf{\dot{z}}$. 

```python
def fire(z0, T, returnfull=False):
    t = np.linspace(0, T, 500)
    thetaseries = odeint(zdot, z0, t)
    if returnfull:
        return thetaseries
    else:
        return thetaseries[-1,:]
```

Define two different functions to measure how we're doing. The first returns the absolute difference between the initial values and the final values produced by our firing function. The convergence function returns wether these two values are within some passed tolerance. 

```python
def measure(theta0, thetaT):
    """Calculate the residual"""
    return thetaT - theta0

def converged(residual, tol):
    """Check convergence to within tol, using the L2 norm"""
    if sum(residual**2) <= tol:
        return True
    else:
        return False
```

Finding the Jacobian matrix, which is augmented by the phase condition. 

```python
def augjac(theta0, T):
    """Compute the augmented jacobian matrix and its inverse.
    
    I'm choosing clarity over efficiency here; if it becomes necessary to change that, I'll refactor."""
    
    # get the baseline residual
    final = fire(theta0, T)
    H0 = measure(theta0, final)
    #print "augjac: H0", H0
    aug0 = np.append(H0, 0.0)
    
    ##Instead append -d_theta_1
    
    # choose a step size that is 0.01% of the residual
    steps = np.abs(H0*0.0001)
    
    jac = []

    # finite differences
    for idx, step in enumerate(steps):
        #print "enumerate", list(enumerate(steps))
        #print "augjac: idx, step", idx, step
        thetaidx0 = np.array(theta0)
        #print "augjac: theta0", thetaidx0
        thetaidx0[idx] += step
        #print "augjac: shifted", thetaidx0
        Hidx = measure(thetaidx0, fire(thetaidx0, T))
        #print "augjac: Hidx", Hidx
        #print "augjac: diff", Hidx-H0
        col = (Hidx-H0)/step
        #print "augjac: col", col
        if idx == 2:
            col = np.append(col,[1.0],0)
        else:
            col = np.append(col,[0.0],0)
        jac.append(col)

    # last column is a little different
    dHdT = zdot(final, T)
    jac.append(np.append(np.array(dHdT), 0.0))
    
    #print "AUGJAC: ", np.matrix(np.array(jac)).T

    # turn it into a matrix and return the inverse
    return np.matrix(np.array(jac).T).I, np.matrix(aug0).T
```

Use the augmented matrix to get our new guess when our initial one fails to converge. 

```python
def getnewtheta(theta0, T):
    aj, ah = augjac(theta0, T)
    deltas = -aj * ah
    #print "NEWTHETA: ", ah
    #print "deltas: ", deltas
    #print "NEWTHETA: ", aj

#    print "NEWTHETA: ", aj.I*deltas
#    print "NEWTHETA: ", ah

    newtheta = np.append(theta0, T)
#    print "NEWTHETA", newtheta
#    print "NEWTHETA", deltas
    return np.reshape(np.array(deltas),5) + np.reshape(newtheta,5)
```

# Section Two: Shooting Method

```python
def shootingmethod(z, T, tol, energies, freqs):
    n = 0


    print "Initial guess:", T, z
    print "Initial energy:", Etot(z)
    
    ts = fire(z, T, returnfull=True)
    converged(z-ts[-1,:], tol)
    
    while not converged(z-ts[-1,:], tol):
        n += 1
        print "Converged: ", converged(z-ts[-1,:], 1.0e-12)
        #print "Residual:", measure(z, ts[-1,:])
        newvec = getnewtheta(z, T)
        T = newvec[-1]
        z = newvec[0:4]
        print "Guess #", n, ":", T, z
        print "Energy #", n, ":", Etot(z)
        ts = fire(z, T, returnfull=True)
        converged(z-ts[-1,:], tol)
        if n > 10:
            print "Convergence loop."
            break
        
    print "Converged: ", converged(z-ts[-1,:], tol)
    print "Residual:", measure(z, ts[-1,:])
    
    figure(1)
    pltres = plt.plot(np.linspace(0,T,500),ts)
    plt.xlabel("Time")
    plt.grid()
        
    energies.append(Etot(z))
    F = (2* np.pi)/T
    freqs.append(F)
    
    print "Energies =", energies
    print "Frequencies =", freqs
    
    figure(2)
    plt.plot(energies, freqs, 'ro')
    plt.ylabel("Frequency")
    plt.xlabel("Energy")
    plt.grid()
```

# Section Three: Run Shooting ($\Delta E$)

This section is for the iterative shooting code stepping through energy, which is useful to find the energy level at which the shooting method first fails to converge (see section four).

```python
def runshooting(nmax, step, evector, eperiod, tol, fullinfo):
    n = 0
    energies = []
    freqs = []
    mult = step
    z = evector*mult
    T = eperiod
    initE = Etot(z)
    print "Initial energy:", initE
    
    while n < nmax: 
        
        n += 1               
        
        #Shooting method code:
        
        if fullinfo == True:
            print "mult", mult
            print "Initial guess:", T, z
            print "Initial energy:", Etot(z)
        
        ts = fire(z, T, returnfull=True)
        converged(z-ts[-1,:], tol)
        m = 0

        while not converged(z-ts[-1,:], tol):
            m += 1
            if fullinfo == True:
                print "Converged: ", converged(z-ts[-1,:], tol)
            #print "Residual:", measure(z, ts[-1,:])
            newvec = getnewtheta(z, T)
            T = newvec[-1]
            z = newvec[0:4]
            if fullinfo == True:
                print "Guess #", m, ":", T, z
                print "Energy #", m, ":", Etot(z)
            ts = fire(z, T, returnfull=True)
            converged(z-ts[-1,:], tol)
        
        if Etot(z) < initE:
            print "Energy dropped on run number", n
            break
        else:
            if fullinfo == True:
                print "Final T =", T, "Final z =", z
                print "Final Energy =", Etot(z)
            
            energies.append(Etot(z))
            F = 1./T
            freqs.append(F)
            mult += step
            z = evector*mult
            T = eperiod

    print "Final Energy:", Etot(z)
    print "Number of runs:", n
    
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xscale("log")
    ax.plot(energies, freqs, 'ro')
    ax.set_ylabel("Frequency")
    ax.set_xlabel("Energy")
    ax.grid()

```

```python
def runshootingjac(nmax, mult, step, evector, eperiod, tol, fullinfo):
    n = 0
    energies = []
    freqs = []
    
    ##pass in some inital energy multiplier based on how much we need to increase the energy before we can start changing
    ##the period. this reduces the jump the method has to make and reduces chances of energy fall out. this multiplier
    ##can be found by hand using the energy step methods in section three --> try to automate later? 
    
    z = evector*mult
    T = eperiod
    initE = Etot(z)
    initF = 1./T
    print "Initial energy:", initE
    print "Iniital frequency:", initF
    
    while n < nmax: 
        n += 1     
        
        if fullinfo == True:
            print "Run Number:", n
            print "Initial guess:", T, z
            print "Initial energy:", Etot(z)
            
        ts = fire(z, T, returnfull=True)
        converged(z-ts[-1,:], tol)

        if not converged(z-ts[-1,:], tol):
            if fullinfo == True:
                print "First nonconvergence. Nonconvergence occured when the period was shifted by", eperiod - T, ". Now using results from previous run regardless of conversion."
                
            while n < nmax:
                
                n += 1
                m = 0
                prevF = 1/T
                
                ts = fire(z, T, returnfull=True)
                converged(z-ts[-1,:], tol)
                
                if fullinfo == True:
                    print "Run Number:", n
                    print "Initial guess:", T, z
                    print "Initial energy:", Etot(z)
                    print "Converged: ", converged(z-ts[-1,:], tol)
                    
                
                
                while not converged(z-ts[-1,:], tol):
                    m += 1
                    if fullinfo == True and m ==1:
                        print "Initial guess:", T, z
                        print "Initial energy:", Etot(z)
                    newvec = getnewthetajac(z, T)
                    T = newvec[-1]
                    z = newvec[0:4]
                    
                    if fullinfo == True:
                        print "Guess #", m, ":", T, z
                        print "Energy #", m, ":", Etot(z)
                        
                    ts = fire(z, T, returnfull=True)
                    converged(z-ts[-1,:], tol)
                    if m > 10:
                        print "Trapped in a convergence loop on run number", n
                        break

                if 1./T < prevF:
                    print "Frequency dropped on run number", n
                    break

                if Etot(z) < initE :
                    print "Energy dropped on run number", n
                    break 
                                
                
                if fullinfo == True:
                    print "Final T =", T, "Final z =", z
                    print "Final Energy =", Etot(z)

                energies.append(Etot(z))
                F = 1/T
                freqs.append(F)
                
                T -= step
                

        else:   
            if fullinfo == True:
                print "Final T =", T, "Final z =", z
                print "Final Energy =", Etot(z)
            
            energies.append(Etot(z))
            F = 1/T
            freqs.append(F)
            
            T -= step
            
            
    print "Final Energy:", Etot(z)
    print "Delta E:", Etot(z) - initE
    print "Final frequency:", 1/T
    print "Delta F:", 1/T - initF
    print "Number of runs:", n
    
    fig = plt.figure()
    fig.set_figwidth(7)
    fig.set_figheight(5)
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xscale("log")
    ax.plot(energies, freqs,'ro')
    ax.set_ylabel("Frequency")
    ax.set_xlabel("Energy")
    ax.grid()

```

```python
def redjac(theta0, T):
    """Compute the reduced jacobian matrix and its inverse.
    Trying to remove various positional variables to try and make the period shift easier. 
    
    I'm choosing clarity over efficiency here; if it becomes necessary to change that, I'll refactor."""
    
    # get the baseline residual
    final = fire(theta0, T)
    H0 = measure(theta0, final)
    #take only the theta_2, d_theta_1, d_theta_2 from baseline residual
    H0 = H0[1:]
    #print "redjac: Residual, H0:", measure(theta0,final), H0
    #append the period onto the baseline residual to create the basis for the reduced jac
    red0 = np.append(H0, 0.0)
    # choose a step size that is 0.01% of the residual
    steps = np.abs(H0*0.0001)
    #print "Steps:", steps
    #create the empty jacobian
    jac = []

    # finite differences
    
    for idx, step in enumerate(steps):
        #print "Enumerate:", list(enumerate(steps))
        #print "redjac: idx, step", idx, step
        #print "redjac: theta0 (original guess)", np.array(theta0)
        
        #take the original guess
        thetaidx0 = np.array(theta0)
        #print "redjac: thetaidx0", thetaidx0
        
        #add the step to the variable in the original guess
        #move forward one extra step to skip over theta_1
        thetaidx0[idx+1] += step
        #print "redjac: shifted", thetaidx0
                
        #take the residual between the original guess and the shifted guess
        #only take theta_2, d_theta_1, d_theta_2
        Hidx = measure(theta0, fire(thetaidx0, T))[1:]
        #print "redjac: Hidx", measure(theta0, fire(thetaidx0, T)), Hidx
        #print "redjac: Diff", Hidx-H0
        
        #take the finite difference
        col = (Hidx-H0)/step
        #print "redjac: col", col
        
        #add the phase condition
        if idx == 1:
            col = np.append(col,[1.0],0)
        else:
            col = np.append(col,[0.0],0)
            
        jac.append(col)
        
    #print np.matrix(np.array(jac)).T
    
    # last column consists of partials taken with respect to time
        
    #print "last column", zdot(final, T), zdot(final, T)[1:]
    dHdT = zdot(final, T)[1:]
    jac.append(np.append(np.array(dHdT), 0.0))
        
    #print "REDJAC: ", np.matrix(np.array(jac).T)
    #print "REDJAC H0:", np.matrix(red0)
    
    # turn it into a matrix and return the inverse
    return np.matrix(np.array(jac).T).I, np.matrix(red0).T
```
