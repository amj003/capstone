This week I worked on my derivation for the eqations of motion for the damped driven double spherical pendulum (DDDSP). I was having problems with the equations of motion coming out incredibly complicated, which were almost impossible to actually work with. To try and better understand what was going on, I went back to the single spherical pendulum. 

I started with the system that is the simplest, an undamped & undriven spherical pendulum. 

Lagrangian:

$$ {1 \over 2} \mathrm{sin}^2(\theta) \left( {d \phi \over dt} \right) ^2 + cos(\theta) + {1 \over 2} \left( {d \theta \over dt} \right) ^2 = 0 $$ 

Equations of Motion: 

$$ {d^2 \phi \over dt^2} = -{ 2 {d \phi \over dt} {d \theta \over dt} \over tan(\theta) } $$

$$ {d^2 \theta \over dt^2} = \mathrm{sin}(\theta) \mathrm{cos}(\theta) \left( {d \phi \over dt} \right) ^2 - \mathrm{sin}(\theta) $$

I then drove the pendulum using the method used in Taylor, rather than adding the cosine term to the definition of the $z$ position. I used the standard driving force $F\mathrm{cos}( \omega t) $.

Lagrangian:

$$ {1 \over 2} \mathrm{sin}^2(\theta) \left( {d \phi \over dt} \right) ^2 + cos(\theta) + {1 \over 2} \left( {d \theta \over dt} \right) ^2 = 0 $$ 

Equations of Motion: 

$$ {d^2 \phi \over dt^2} = -{ 2 {d \phi \over dt} {d \theta \over dt} \over tan(\theta) } $$

$$ {d^2 \theta \over dt^2} = F \mathrm{cos}(t \omega) + \mathrm{sin}(\theta) \mathrm{cos}(\theta) \left( {d \phi \over dt} \right) ^2 - \mathrm{sin}(\theta) $$

I then added a damped force according to Goldstein's Rayleigh dissipation function. 


Lagrangian:

$$ {1 \over 2} \mathrm{sin}^2(\theta) \left( {d \phi \over dt} \right) ^2 + cos(\theta) + {1 \over 2} \left( {d \theta \over dt} \right) ^2 = 0 $$ 

Equations of Motion: 

$$ {d^2 \phi \over dt^2} = -\left( b + { 2  {d \theta \over dt} \over tan(\theta) } \right) {d \phi \over dt}$$

$$ {d^2 \theta \over dt^2} = F \mathrm{cos}(t \omega) - b {d \theta \over dt} + \mathrm{sin}(\theta) \mathrm{cos}(\theta) \left( {d \phi \over dt} \right) ^2 - \mathrm{sin}(\theta) $$
