```python
import numpy as np
from sympy import *
from sympy.physics.vector import dynamicsymbols
init_printing()

#Let l = 1, m = 1, g = 1 
```

Goal: Find the Lagrangian for a double spherical pendulum.

The first step is to find the single spherical pendulum Lagrangian. The two degrees of freedom are $\phi_1$ and $\theta_1$. The cartesian coordinates of the mass of the pendulum is defined in terms of these two variables. 

```python
#Use dynamicsymbols to make theta_1 and phi_1 depend on time.
theta_1 = dynamicsymbols('theta_1')
phi_1 = dynamicsymbols('phi_1')
d_theta_1 = dynamicsymbols('d_theta_1}')
d_phi_1 = dynamicsymbols('d_phi_1')
t = Symbol('t')
x_1 = Symbol('x_1')
y_1 = Symbol('y_1')
z_1 = Symbol('z_1')
d_x_1 = Symbol('d_x_1')
d_y_1 = Symbol('d_y_1')
d_z_1 = Symbol('d_z_1')
L = Symbol('L')
theta_2 = dynamicsymbols('theta_2')
phi_2 = dynamicsymbols('phi_2')
x_2 = Symbol('x_2')
y_2 = Symbol('y_2')
z_2 = Symbol('z_2')
d_x_2 = Symbol('d_x_2')
d_y_2 = Symbol('d_y_2')
d_z_2 = Symbol('d_z_2')
w = Symbol('w')
F = Symbol('F')
```

```python
x_1 = sin(theta_1)*cos(phi_1)
y_1 = sin(theta_1)*sin(phi_1)
z_1 = -cos(theta_1) 
#+ F*cos(w*t)

#Define d_theta_1 and d_phi_1

d_theta_1 = diff(theta_1, Symbol('t'))
d_phi_1 = diff(phi_1, Symbol('t'))

#Define the derivatives of the coordinate system 

d_x_1 = diff(x_1,Symbol('t'))
d_y_1 = diff(y_1,Symbol('t'))
d_z_1 = diff(z_1,Symbol('t'))

#Find the velocity of the pendulum

v_1 = simplify(expand((d_x_1)**2 + (d_y_1)**2 + (d_z_1)**2))

#Find kinetic energy and potential energy. Kinetic is T = 1/2*m*v**2 and 
#potential is U = mgh where h is just z_1.
#The driving force is already in z_1.
#Define the lagrangian.

T = 0.5*v_1
U = z_1
L_1 = T - U

L_1
```

```python
x_2 = sin(theta_1)*cos(phi_1) + sin(theta_2)*cos(phi_2)
y_2 = sin(theta_1)*sin(phi_1) + sin(theta_2)*sin(phi_2)
z_2 = -cos(theta_1) - cos(theta_2) 
#+ F*cos(w*t)

d_x_2 = diff(x_2,Symbol('t'))
d_y_2 = diff(y_2,Symbol('t'))
d_z_2 = diff(z_2,Symbol('t'))

d_theta_2 = diff(theta_2,Symbol('t'))
d_phi_2 = diff(phi_2,Symbol('t'))

v_2 = simplify(expand((d_x_2)**2 + (d_y_2)**2 + (d_z_2)**2))
T = 0.5*(v_1 + v_2)
U = z_1 +  z_2
L_2 = T - U

simplify(expand(L_2))
```

```python
L_d_theta_1 = diff(L, theta_1)

L_dd_theta_1 = diff(L, d_theta_1)
eq_theta_1 = simplify(diff(L_dd_theta_1, Symbol('t')) - L_d_theta_1)
```

```python

L_d_theta_2 = diff(L, theta_2)
L_dd_theta_2 = diff(L, d_theta_2)

eq_theta_2 = simplify(diff(L_dd_theta_2, Symbol('t')) - L_d_theta_2)

```

```python
L_d_phi_1 = diff(L, phi_1)
L_dd_phi_1 = diff(L, d_phi_1)

eq_phi_1 = simplify(expand(diff(L_dd_phi_1, Symbol('t')) - L_d_phi_1))
```

```python
L_d_phi_2 = diff(L, phi_2)
L_dd_phi_2 = diff(L, d_phi_2)

eq_phi_2 = simplify(expand(diff(L_dd_phi_2,Symbol('t')) - L_d_phi_2))
```

```python
b = Symbol('b')

scrF = 0.5*b*v_2

scrF_theta_1 = diff(scrF, d_theta_1)
scrF_theta_2 = diff(scrF, d_theta_2)
scrF_phi_1 = diff(scrF, d_phi_1)
scrF_phi_2 = diff(scrF, d_phi_2)

```

```python
c_eq_theta_1 = eq_theta_1 + scrF_theta_1
c_eq_theta_2 = eq_theta_2 + scrF_theta_2
c_eq_phi_1 = eq_phi_1 + scrF_phi_1
c_eq_phi_2 = eq_phi_2 + scrF_phi_2

```

```python
solve([c_eq_theta_1, c_eq_theta_2, c_eq_phi_1, c_eq_phi_2], [diff(d_theta_1, Symbol('t')), diff(d_theta_2, Symbol('t')), diff(d_phi_1, Symbol('t')), diff(d_phi_2, Symbol('t'))])

```
