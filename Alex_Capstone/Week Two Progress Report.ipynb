### Damped Driven Pendulum

From Taylor <i>Classical Mechanics</i> we know that the equation for the damped driven pendulum is:

$$ \ddot{\phi} + 2 \beta \dot{\phi} + \omega_0^2 \mathrm{sin} \phi = \gamma \omega_0^2 \mathrm{cos} ( \omega t) $$

Where $\beta$ is a damping constant, $\gamma$ is a driving force constant, $\omega_0$ is the natural frequency of the pendulum, and $\omega$ is the driving frequency.

This equation comes out of the Lagrangian method. To investigate the motion of the pendulum we can use a RK method to model the angular position and angular velocity based on initial conditions. 

One option is to write a Runge Kutta method.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy import arange

%pylab inline
```

```python
#phi'' = -2B*phi' - w_0**2*sin(phi) + gamma*w_0**2*cos(wt)

#u1(t) = phi(t)
#u2(t) = phi'(t)
#u1'(t) = u2(t)
#u2'(t) = -2B*phi' - w_0**2*sin(phi) + gamma*w_0**2*cos(wt) 
#u2'(t) = -2*B*u2(t) - w_0**2*sin(u1(t)) + gamma*w_0**2*cos(wt)

def RK(B,w_0,w,gamma, timemax):
    h = 1
    def f1(t, u1, u2):
        return u2
    def f2(t, u1, u2):
        return -2*B*u2 - w_0**2*sin(u1) + gamma*w_0**2*cos(w*t)
    u1_0 = .5
    u2_0 = 1.
    t0 = 0.0
    n = 0
    time = arange(0, timemax, h)
    list_w1 = []
    list_w2 = []
    while n == 0:
        k1_1 = h*f1(t0, u1_0, u2_0)
        k1_2 = h*f2(t0, u1_0, u2_0)
        k2_1 = h*f1(t0 + 0.5*h, u1_0 + 0.5*k1_1, u2_0 + 0.5*k1_2)
        k2_2 = h*f2(t0 + 0.5*h, u1_0 + 0.5*k1_1, u2_0 + 0.5*k1_2)
        k3_1 = h*f1(t0 + 0.5*h, u1_0 + 0.5*k2_1, u2_0 + 0.5*k2_2)
        k3_2 = h*f2(t0 + 0.5*h, u1_0 + 0.5*k2_1, u2_0 + 0.5*k2_2)
        k4_1 = h*f1(t0 + h, u1_0 + k3_1, u2_0 + k3_2)
        k4_2 = h*f2(t0 + h, u1_0 + k3_1, u2_0 + k3_2)
        w1_1 = u1_0 + (1.0/6.0)*(k1_1 + 2*k2_1 + 2*k3_1 + k4_1)
        w2_1 = u2_0 + (1.0/6.0)*(k1_2 + 2*k2_2 + 2*k3_2 + k4_2)
        u1_0 = w1_1
        u2_0 = w2_1
        t0 = t0 + h
        list_w1.append(w1_1)
        list_w2.append(w2_1)
        if t0 == time[len(time)-1] + h :
            n = 1
        else:
            n = 0
    w1 = np.array(list_w1)
    w2 = np.array(list_w2)
    figure(1, figsize = (10,10))
    label = " Beta = " + str(B) + ", Gamma =" + str(gamma)+", Driving Frequency = "+str(w_0)
    f1 = scatter(w1,w2,label = label)
    plt.legend(fontsize =12)
    plt.xlabel('Angular Position',fontsize = 15)
    plt.ylabel('Angular Velocity',fontsize = 15)
    figure(2, figsize = (10,10))
    f2 = plot(time, w1, label = 'Angular Position')
    f3 = plot(time, w2, label = 'Angular Velocity')
    plt.xlabel('Time',fontsize = 15)
    plt.legend(fontsize = 12)
    
```

So after breaking this method, I moved on to using scipy.integrate. Integrate is a module for Python with useful methods for solving differential equations. 

```python
from scipy.integrate import odeint

#phi'' = -2B*phi' - w_0**2*sin(phi) + gamma*w_0**2*cos(wt)

#u1(t) = phi(t)
#u2(t) = phi'(t)
#u1'(t) = u2(t)
#u2'(t) = -2B*phi' - w_0**2*sin(phi) + gamma*w_0**2*cos(wt) 
#u2'(t) = -2*B*u2(t) - w_0**2*sin(u1(t)) + gamma*w_0**2*cos(wt)

def sciode(B, w_0, w, gamma, timemax,timestep):
    def f(x_vec, t):
        u1, u2 = x_vec
        return [u2, -2*B*u2 - w_0**2*sin(u1) + gamma*w_0**2*cos(w*t)]

    x_0 = [0.5, 1.0]
    time = arange(0, timemax, timestep)
    results = odeint(f, x_0, time)
        
    theta = list(results[n,0] for n in time)
    theta_dot = list(results[n,1] for n in time)
    figure(1, figsize = (15,10))
    label = " Beta = " + str(B) + ", Gamma =" + str(gamma)+", Driving Frequency = "+str(w_0)
    f1 = scatter(theta,theta_dot, label = label)
    plt.legend()
    plt.xlabel('Angular Position',fontsize = 15)
    plt.ylabel('Angular Velocity',fontsize = 15)
    figure(2, figsize = (15,10))
    f2 = plot(time, theta, label = 'Angular Position')
    f3 = plot(time, theta_dot, label = 'Angular Velocity')
    plt.xlabel('Time',fontsize = 15)
    plt.legend(fontsize = 12)
```

We can test this method with a few known examples. If the damping is large and there is no driving force, the pendulum should slow to a stop:

```python
sciode(1,3,1,0, 600, .01)
```

On the other hand, if the driving force is large and damping is small, the pendulum should continue along with the driven motion. 

```python
sciode(0, 1, 1, 8, 600, .1)
```

If there is no damping or driving, the pendulum should continue with its own motion forever. 

```python
sciode(0, 10, 0, 0, 600, .01)
```

First I need to work out the DDSP equations of motion with a driving force and a damping force. The driving force can be folded into the gravitational potential energy when formulating the lagrangian:
$$L = T - (U_0 + U_{driven})$$
$$U_0 = mgy$$
$$U_{driven} = mg*c_{proportionality}*cos(t)$$

However, the damping force will need to be added to the equations of motion, using <i> Rayleigh's dissipation function </i> $\mathscr{F}$ (Goldstein <i>Classical Mechanics</i> pg. 24), which allows for the addition of a velocity-dependent non-conservative force. This will have the form:

$${ d \over dt} {\partial L \over \partial \dot{q_i}} - {\partial L \over \partial q_i} + {\partial \mathscr{F} \over \partial \dot{q_i}} = 0$$
