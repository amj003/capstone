To fix the problem with the direction of the driving force a cosine term was added:

Lagrangian:

$$ {1 \over 2} \mathrm{sin}^2(\theta) \left( {d \phi \over dt} \right) ^2 + cos(\theta) + {1 \over 2} \left( {d \theta \over dt} \right) ^2 = 0 $$ 

Equations of Motion: 

$$ {d^2 \phi \over dt^2} = -{ 2 {d \phi \over dt} {d \theta \over dt} \over tan(\theta) } $$

$$ {d^2 \theta \over dt^2} = F \mathrm{cos}(t \omega)*\mathrm{cos}(\theta) + \mathrm{sin}(\theta) \mathrm{cos}(\theta) \left( {d \phi \over dt} \right) ^2 - \mathrm{sin}(\theta) $$



I also added this to the damped equations of motion. 

Lagrangian:

$$ {1 \over 2} \mathrm{sin}^2(\theta) \left( {d \phi \over dt} \right) ^2 + cos(\theta) + {1 \over 2} \left( {d \theta \over dt} \right) ^2 = 0 $$ 

Equations of Motion: 

$$ {d^2 \phi \over dt^2} = -\left( b + { 2  {d \theta \over dt} \over tan(\theta) } \right) {d \phi \over dt}$$

$$ {d^2 \theta \over dt^2} = F \mathrm{cos}(t \omega)*\mathrm{cos}(\theta) - b {d \theta \over dt} + \mathrm{sin}(\theta) \mathrm{cos}(\theta) \left( {d \phi \over dt} \right) ^2 - \mathrm{sin}(\theta) $$
